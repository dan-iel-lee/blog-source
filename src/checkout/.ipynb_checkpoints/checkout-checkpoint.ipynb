```python
fig, axs = plt.subplots(1, math.floor((len(dp)-2)/3), figsize = (20, 5), sharex='col')
for i in range(1, math.floor(len(dp)/3)+1):
    L = i * 3
    ys = dp[L]
    xs = range(0, len(dp[L]))
    ax = axs[i-1]
    ax.scatter(xs, ys)
    ax.set_title("L=" + str(L))
    ax.set_xlabel("N (total cart count)")

fig.suptitle("Expected lane length vs N for different choices of L", fontsize=16)
fig.text(0.04, 0.5, 'Expected lane length', va='center', rotation='vertical')
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 6))
# first 5
for j in range(0, round(len(dp[0])/40)):
    N = 20 * j
    ys = [dp[i][N] for i in range(0, len(dp))]
    xs = range(0, len(dp))
    axs[0].scatter(xs, ys, label="N=" + str(N))

# last 5
for j in range(round(len(dp[0])/40), math.floor(len(dp[0])/20)):
    N = 20 * j
    ys = [dp[i][N] for i in range(0, len(dp))]
    xs = range(0, len(dp))
    axs[1].scatter(xs, ys, label="N=" + str(N))
    
axs[0].legend()
axs[1].legend()
fig.suptitle("Expected lane length vs L for different choices of N", fontsize=16)
plt.show()
```
